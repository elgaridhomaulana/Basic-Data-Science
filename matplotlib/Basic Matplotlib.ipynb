```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [20, 30, 40, 50, 60, 70, 80]
y = [5000, 6500, 8000, 5000, 4500, 4000, 4000]
```

```python
plt.scatter(x,y, s = 60, c = 'green', marker='^')
plt.title('Relationship between age and Revenue')
plt.xlabel('Age')
plt.ylabel('Revenue')
plt.xlim(0, 100)
plt.ylim(0, 10000)
plt.show()
```

```python
x = range(6)
plt.plot(x, [xi**2 for xi in x])
plt.show()
```

```python
x = np.arange(0.0, 6.0, 0.01)
plt.plot(x, [xi**2 for xi in x])
plt.show()
```

```python
x = range(1, 5)
plt.plot(x, [xi*1.5 for xi in x])
plt.plot(x, [xi*3 for xi in x])
plt.plot(x, [xi/2 for xi in x])
plt.show()
```

```python
x = range(1, 5)
plt.plot(x, [xi*1.5 for xi in x], x, [xi*3 for xi in x], x, [xi/2 for xi in x])
plt.show()
```

```python
x = np.arange(1, 5)
plt.plot(x, x*1.5, x, x*3, x, x/3)
plt.show()
```

```python
x = np.arange(0, 5)
plt.plot(x, x*1.5, x, x*3, x, x/3)
plt.axis([0, 6, 0, 13])
plt.grid(True)
plt.show()
```

```python
plt.plot([1,3,2,4])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title of the Graph')
plt.show()
```

```python
x = np.arange(1,5)
plt.plot(x, x*1.5, label = 'Normal')
plt.plot(x, x*3, label = 'Fast')
plt.plot(x, x/1.5, label = 'Slow')
plt.legend()
plt.show()
```

```python
x = np.arange(1,5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x/3, label='Slow')
plt.plot(x, x*3, label='Fast')
plt.grid(True)
plt.title('Growth Measurement')
plt.xlabel('Sample')
plt.ylabel('Values')
plt.legend(loc = 'upper left')
plt.savefig('Growth.png')
plt.show()
```

```python
y = np.arange(1,3)
plt.plot(y, 'y')
plt.plot(y+1, 'b')
plt.plot(y+2, 'c')
plt.show()
```

```python
y = np.arange(1,3)
plt.plot(y, 'y', y+1, 'b', y+2, 'c')
plt.show()
```

```python
y = np.arange(1,3)
plt.plot(y, '--', y+1, '-.', y+2, ':')
plt.show()
```

```python
y = np.arange(1,3,0.2)
# x cross marker
# o circle marker
# D diamond marker
# ^ Triangle up marker
# s Square marker

plt.plot(y, 'x', y+0.5, 'o', y+1, 'D', y+1.5, '^', y+2, 's')
plt.show()
```

```python
y = np.arange(1,3,0.3)
plt.plot(y, 'cx--', y+1, 'mo:', y+2, 'kp-.')
plt.show()
```

```python
y = np.arange(1,3,0.3)
plt.plot(y, color='blue', linestyle='dashdot', linewidth=3, marker='^', 
         markerfacecolor='white', markeredgecolor = 'yellow', markeredgewidth=3, markersize=15)
plt.show()
```

```python
x = [5, 3, 7, 2, 4, 1]
plt.plot(x)
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f'])
plt.yticks(range(1,8,2))
plt.show()
```

### Type of Plots

Histogram Plot

```python
y = np.random.randn(1000)
plt.hist(y, edgecolor='black')
plt.show()
```

```python
y = np.random.randn(1000)
plt.hist(y, edgecolor='black', bins=30)
plt.show()
```

Error Bar Chart

```python
x = np.arange(0, 4, 0.2)
y = np.exp(-x)
e1 = 0.1 * np.abs(np.random.randn(len(y)))
plt.errorbar(x,y, yerr=e1, fmt='.-')
plt.show()
```

Bar Chart

```python
plt.bar([1,2,3], [3, 2, 5])
plt.show()
```

```python
dictionary = {'A':20, 'B':30, 'C':10, 'D':20, 'E':40}
for element, key in enumerate(dictionary):
    plt.bar(element, dictionary[key])

plt.xticks(np.arange(len(dictionary)), dictionary.keys())
plt.show()
```

Pie Charts

```python
plt.figure(figsize=(4,4))
x = [45, 30, 25]
labels = ['Woman', 'Man', 'Children']
plt.pie(x, labels=labels)
plt.show()
```

```python
x = [4, 9, 21, 55, 30, 18]
labels = ['Swiss', 'Austria', 'Spain', 'Italy', 'France', 'Cyprus']
explode = [0.2, 0.1, 0, 0, 0.1, 0]
plt.pie(x, labels=labels, explode=explode, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y, edgecolors='black')
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
size = 50*np.random.randn(1000)

plt.scatter(x,y, s=size, edgecolors='black')
plt.show()
```
