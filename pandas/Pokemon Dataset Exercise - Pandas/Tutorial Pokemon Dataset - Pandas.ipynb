### Sorting Data

### Making Some Change To The DataFrame

```python
pokemon_data['Total'] = pokemon_data['HP'] + pokemon_data['Attack'] + pokemon_data['Defense'] + pokemon_data['Sp. Atk'] + pokemon_data['Sp. Def'] + pokemon_data['Speed']
```

```python
pokemon_data[0:5]
```

```python
#Changing the order of the columns
cols = list(pokemon_data.columns.values)
pokemon_data = pokemon_data[cols[0:10] + [cols[-1]] + cols[10:12]]
```

```python
pokemon_data
```

```python
pokemon_data.to_csv('modified_pokemon_data.csv', index=False)
```

### Filtering Data

```python
# so we can called the dataframe more easier :)
df = pokemon_data
```

```python
#Let's say we want to find the type 1 fire and legendary pokemon
df.loc[(df['Type 1'] == 'Fire') & (df['Legendary'] == True)]
```

We could see that there are Type 1 = Fire which are also legendary pokemons

```python
# Find Name that contain Mega
df.loc[df['Name'].str.contains('Mega')]
```

```python
#Filtering using regular expressions
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```
